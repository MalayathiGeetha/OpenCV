```python
#Edge Detection
import cv2
from google.colab.patches import cv2_imshow

img=cv2.imread("boy.jpeg")
img=cv2.resize(img,(500,500))

new_img=cv2.Canny(img,200,200,apertureSize=5,L2gradient=True)


cv2_imshow(new_img)
```

```python
#Image Scaling Rotating
import numpy as np

import cv2

org_img=cv2.imread("boy.jpeg")
res_img=cv2.resize(org_img,(500,500))
w,h=res_img.shape[0],res_img.shape[1]
m=cv2.getRotationMatrix2D((w/2,h/2),70,1)


new_img=cv2.warpAffine(res_img,m,(w,h))
h=np.hstack((res_img,new_img))
#cv2_imshow(new_img)
cv2_imshow(h)
```

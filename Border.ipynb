```python
#Make Border
import cv2
import numpy as np
from google.colab import files
uploaded = files.upload()
```

```python
from google.colab.patches import cv2_imshow
img=cv2.imread("abdul kalam.jpeg")
img=cv2.resize(img,(300,300))

img1=cv2.copyMakeBorder(img,20,20,20,20,cv2.BORDER_CONSTANT,None,value=2)
img1=cv2.resize(img1,(300,300))

img2=cv2.copyMakeBorder(img,20,20,20,20,cv2.BORDER_DEFAULT,None,value=2)
img2=cv2.resize(img2,(300,300))

#REFLECT101
img3=cv2.copyMakeBorder(img,20,20,20,20,cv2.BORDER_REPLICATE,None,value=2)
img3=cv2.resize(img3,(300,300))
h=np.hstack((img,img1,img2,img3))
cv2_imshow(h)
```

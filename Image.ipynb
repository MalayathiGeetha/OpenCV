```python
import cv2
from google.colab.patches import cv2_imshow

old_img=cv2.imread("show.jpeg")
old_img=cv2.resize(old_img,(500,700))


new_img=cv2.rectangle(img=old_img,pt1=(120,200),pt2=(400,500),color=(0,255,0),thickness=4,lineType=16)

# cv2_imshow(old_img)
cv2_imshow(new_img)
```

```python
old_img=cv2.imread("abdul kalam.jpeg")
old_img=cv2.resize(old_img,(500,600))

new_img=cv2.circle(img=old_img,center=(300,200),radius=170,color=(0,255,0),thickness=4,lineType=16)
cv2_imshow(new_img)
```

```python
#eclipse also same
#polygon
import numpy as np

old_img=cv2.imread("abdul kalam.jpeg")
old_img=cv2.resize(old_img,(500,600))


new_img=cv2.polylines(img=old_img,pts=[np.array([[100,200],[150,100],[300,200],[300,500],[100,500]])],isClosed=True,color=(0,255,0),thickness=4,lineType=16)
cv2_imshow(new_img)
```

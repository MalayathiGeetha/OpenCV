```python
import numpy as np

img = cv2.imread("img1.jpg")
resize=cv2.resize(img,(300,300))
h=np.hstack((resize,resize))
v=np.vstack((resize,resize))
cv2_imshow(h)
print("/n")
cv2_imshow(v)

print("/n")
v=np.vstack((h,h))
cv2_imshow(v)
```

```python

img_get=cv2.imread("img1.jpg")
img_get=cv2.resize(img_get,(500,600))


txt=cv2.putText(
    img=img_get,
text="wscube",
org=(50,80),
fontFace=cv2.FONT_HERSHEY_DUPLEX,
fontScale=3,
color=(0,0,255),
thickness=3,
lineType=cv2.LINE_8,
bottomLeftOrigin=False
)

cv2_imshow(img_get)
```
